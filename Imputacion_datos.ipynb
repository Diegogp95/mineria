```python
import numpy as np
import pandas as pd
import numpy as np
```

```python
df_data = pd.read_csv('data_2020.csv')
df_data['codigo_estacion'] = pd.to_datetime(
    df_data['codigo_estacion'].astype(str),
    format='%Y-%m-%d')
df_data
```

```python
df_data = df_data.fillna(-9999.0)
df_data.isna().describe()
```

```python
df = df_data
df
```

```python
#imputacion media -> No funciona porq solo es para datos numéricos
import pandas as pd
from sklearn.impute import SimpleImputer
imputador = SimpleImputer(strategy='mean')
imputador.fit(df)
df_imputado = pd.DataFrame(imputador.transform(df), columns=df.columns)
df_imputado
```

```python
#imputacion valor mas frecuente
import pandas as pd
from sklearn.impute import SimpleImputer
imputador = SimpleImputer(strategy='most_frequent')
imputador.fit(df)
df_imputado = pd.DataFrame(imputador.transform(df), columns=df.columns)
df_imputado
```

```python
#Valor específico
imputador = SimpleImputer(strategy='constant', fill_value='Desconocido')
imputador.fit(df)
df_imputado2 = pd.DataFrame(imputador.transform(df), columns=df.columns)
df_imputado2
```

```python
#Interpolacion
columnas_numericas = df.select_dtypes(include=[np.number]).columns
df_imputado3 = df[columnas_numericas].interpolate(method='linear')
df_imputado3
```

```python
#Basada en modelos
from sklearn.impute import KNNImputer

#columnas categóricas para la codificación one-hot
columnas_categoricas = df.select_dtypes(include=['object']).columns

#codificación one-hot a columnas categóricas
df_encoded = pd.get_dummies(df, columns=columnas_categoricas)

#imputación con KNN
imputador = KNNImputer(n_neighbors=5)
df_imputado4 = pd.DataFrame(imputador.fit_transform(df_encoded), columns=df_encoded.columns)
df_imputado4
```
