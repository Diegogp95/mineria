```python
import pandas as pd

# Ruta del archivo de texto (.txt)
archivo_txt = "cr2_prDaily_2020_stations_ghcn.txt"

# Leer el archivo de texto
with open(archivo_txt, 'r', encoding='utf-8') as file:
    contenido = file.readlines()

# Limpiar y formatear los datos
contenido = [line.strip().split(',') for line in contenido]

# Convertir a un DataFrame de pandas
df = pd.DataFrame(contenido)

# Guardar el DataFrame como archivo CSV
archivo_csv = "cr2_prDaily_2020_stations_ghcn.csv"
df.to_csv(archivo_csv, index=False)
```

```python
df = pd.read_csv("cr2_prDaily_2020_stations_ghcn.csv", header=0)

# Asignar la primera fila como encabezado
df.columns = df.iloc[0]

# Eliminar la primera fila (antiguo encabezado)
df = df[1:]

# Reiniciar los √≠ndices del DataFrame
df.reset_index(drop=True, inplace=True)

df
```

```python
df = df.dropna()
df
```

```python
#imputacion media -> No funciona porq solo es para datos num√©ricos
import pandas as pd
from sklearn.impute import SimpleImputer
imputador = SimpleImputer(strategy='mean')
imputador.fit(df)
df_imputado = pd.DataFrame(imputador.transform(df), columns=df.columns)
df_imputado
```

```python
#imputacion valor mas frecuente
import pandas as pd
from sklearn.impute import SimpleImputer
imputador = SimpleImputer(strategy='most_frequent')
imputador.fit(df)
df_imputado = pd.DataFrame(imputador.transform(df), columns=df.columns)
df_imputado
```

```python
#Valor espec√≠fico
imputador = SimpleImputer(strategy='constant', fill_value='Desconocido')
imputador.fit(df)
df_imputado2 = pd.DataFrame(imputador.transform(df), columns=df.columns)
df_imputado2
```

```python
#Interpolacion
columnas_numericas = df.select_dtypes(include=[np.number]).columns
df_imputado3 = df[columnas_numericas].interpolate(method='linear')
df_imputado3
```

```python
#Basada en modelos
from sklearn.impute import KNNImputer

#columnas categ√≥ricas para la codificaci√≥n one-hot
columnas_categoricas = df.select_dtypes(include=['object']).columns

#codificaci√≥n one-hot a columnas categ√≥ricas
df_encoded = pd.get_dummies(df, columns=columnas_categoricas)

#imputaci√≥n con KNN
imputador = KNNImputer(n_neighbors=5)
df_imputado4 = pd.DataFrame(imputador.fit_transform(df_encoded), columns=df_encoded.columns)
df_imputado4
```
